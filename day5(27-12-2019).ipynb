```
#manipulating a list
#1.adding elements
list1=[34,45,56,67]
print("before adding elements:",list1)
list1.append(100) #accepts single value
list1.append(200)
list1.append(300)
print("after adding elements:",list1)
```

#**dictionary**:
 #syntax: {key1:value1,key2:value2,...}
 #key:

1.   key should be unique
2.   key should be immutable-->int,float,string,tuple

```
dict1={"name":"percy","age":"17","address":"hyd"}
dict1["address"]
```

```
#replace values
dict1["age"]="15"
print(dict1)
```
